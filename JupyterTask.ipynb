## Demonstration of _python_ and _matplotlib_

````python
a = 5
b = 7
````

````python
c = a + b
c
````

````python
import matplotlib.pyplot as plt
````

````python
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]
````

````python
plt.plot(x, y, marker="o")
plt.title("Simple Line Plot: y = x^2")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.show()
````
